```python
import matplotlib.pyplot as plt
%matplotlib inline

#sin커브
x = np.arange(0,10,0.1)
y = np.sin(x)

plt.plot(x,y)
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('sin and cos')
plt.legend(['sin','cos'])

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('sin')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cos')

plt.show
```
